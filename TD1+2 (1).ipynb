## 1. Contexte général

La conduite automobile est une activité complexe qui mobilise plusieurs ressources cognitives : perception visuelle, attention soutenue, prise de décision, anticipation, contrôle moteur et gestion des imprévus.

Le papier CL-Drive s'intéresse donc à une question centrale :

> Peut-on estimer automatiquement la charge cognitive d'un conducteur à partir de signaux physiologiques et comportementaux mesurés pendant la conduite ?

#### Q1. Pourquoi est-il intéressant d'estimer la charge cognitive d'un conducteur ?

**Réponse :**

Parce qu'une charge cognitive trop élevée dégrade les performances de conduite (temps de réaction plus long, erreurs de trajectoire, mauvaise anticipation). L'objectif final est de réduire le nombre d'accidents liés à la distraction ou à la surcharge mentale.

## 2. Définition : charge cognitive

#### Q2. Comment définir la charge cognitive ?

**Réponse :**

C'est la quantité de ressources mentales mobilisées par la mémoire de travail pour accomplir une tâche. Au-delà d'un certain seuil les performances chutent.

#### Q3. Dans le contexte de la conduite, de quels facteurs la charge cognitive peut-elle dépendre ?

**Réponse :**

De la difficulté intrinsèque de la tâche de conduite (route étroite, nuit, neige, virages serrés) mais aussi de facteurs externes comme une conversation téléphonique/une navigation GPS complexe/un état émotionnel particulier. 

Dans le papier, la charge cognitive est évaluée à l'aide d'une échelle subjective appelée **PAAS**, allant de 1 à 9.

## 5. Modalités mesurées

Le dataset CL-Drive est **multimodal**. Il contient plusieurs types de signaux enregistrés pendant la même tâche de conduite.

#### Q7. Compléter le tableau suivant.

| Modalité | Signification | Signal mesuré | Rôle possible |
|---|---|---|---|
| EEG |Électroencéphalogramme | Activité électrique du cerveau |Modalité principale: reflète directement l'état cognitif|
| ECG | Électrocardiogramme | Activité électrique du cœur |Mesure la variabilité du rythme cardiaque, liée au stress et à la charge|
| EDA | Activité électrodermale | Conductance de la peau (sudation) |Reflète l'activation du système nerveux sympathique|
| Gaze | Oculométrie | Mouvements oculaires, diamètre pupillaire, clignements |Indicateur comportemental de la charge attentionnelle|

Dans le projet, la modalité principale sera l’**EEG**. Les autres modalités serviront éventuellement d’extensions pour discuter de la fusion multimodale.

#### Q8. Pourquoi peut-il être intéressant de combiner EEG, ECG, EDA et Gaze pour estimer la charge cognitive ?

**Réponse :**

Chaque modalité capte un aspect différent de la charge cognitive. En les combinant on obtient une image plus complète et robuste. Les résultats du papier confirment que les modèles multimodaux font mieux que les modèles unimodaux dans tous les scénarios testés.

## 7. Protocole expérimental

#### Q11. Quelle est la différence entre un **protocole expérimental** et un **dataset exploitable en machine learning** ?

**Réponse :**

Le protocole expérimental décrit comment les données ont été collectées : quels capteurs?quelles consignes? quel ordre des tâches? comment les participants ont été recrutés?. Le dataset exploitable en ML est le résultat concret sous forme des fichiers de signaux alignés avec des labels et qui sont prêts à être découpés en fenêtres et injectés dans un modèle.

#### Q12. Donner une description détaillée du protocole expérimental utilisé.

**Réponse :**

Chaque participant (consent) arrive et se fait installer les capteurs (casque EEG Muse, Shimmer ECG et EDA, lunettes oculométriques Tobii). On enregistre d'abord 3 minutes de baseline (signal au repos). Ensuite le participant fait le scénario d'orientation (autoroute simple) pour s'habituer au simulateur. Puis pour chacun des 9 scénarios: 2 minutes de baseline, le scénario de 3 minutes, 2 minutes de repos. Pendant les 3 minutes du scénario un signal sonore toutes les 10 secondes invite le participant à annoncer verbalement son score PAAS. Un membre de l'équipe note les scores. En tout on a 18 scores par scénario.

#### Q13. Pourquoi les auteurs collectent-ils une baseline avant les scénarios ?

**Réponse :**

Pour avoir un signal de référence propre à chaque participant et reduire la variabilité intersujets. Les signaux physiologiques varient beaucoup d'une personne à l'autre. 

#### Q14. Pourquoi prévoir un scénario d’orientation avant les scénarios principaux ?

**Réponse :**

Pour que le participant comprenne le fonctionnement du simulateur avant que les vrais scénarios commencent. Sans ça, les premiers enregistrements seraient biaisés par la phase d'apprentissage du simulateur lui-même, pas par la charge cognitive de conduite.

#### Q15. Pourquoi prévoir des périodes de repos entre les scénarios ?

### Réponse

Pour 2 raisons. D'abord, permettre au participant de revenir à un état de base avant le scénario suivant, sinon la charge résiduelle du scénario précédent contaminerait les données. Ensuite, réduire le risque de Simulation Adaptation Syndrome (SAS) qui peut provoquer malaise ou nausées ou vertiges.


#### Q16. Pourquoi demander un score de charge cognitive toutes les 10 secondes plutôt qu’une seule fois à la fin du scénario ?

### Réponse

La charge cognitive évolue pendant un scénario donc un score unique à la fin est une moyenne mémorielle biaisée. En demandant un score toutes les 10 secondes on obtient une mesure dense du signal physiologique, ce qui permet de créer des paires (segment,label) précises et exploitables en ML

## 8. Scénarios de conduite

Le dataset contient 9 scénarios principaux, conçus pour induire différents niveaux de charge cognitive.

#### Q17. Compléter le tableau suivant.

| Scénario | Description synthétique |
|---:|---|
| 1 | Autoroute de jour, maintenir 80 km/h |
| 2 | Conduite de nuit à 80 km/h |
| 3 | Conduite de nuit sur autoroute avec neige à 80 km/h |
| 4 | Toucher une balle de tennis avec le pneu, maintenir la précision, accélérer |
| 5 | Slalom entre des portes, précision + accélération |
| 6 | Passage dans un couloir étroit, précision + accélération |
| 7 | Virages à 90 degrés gauche/droite, précision + accélération |
| 8 | Demi tour en 3 points en suivant des instructions |
| 9 | Navigation dans une ruelle très étroite, précision + accélération |

L'hypothèse générale est que les scénarios les plus complexes induisent une charge cognitive plus élevée.

#### Q18. Pourquoi utiliser plusieurs scénarios de conduite au lieu d'un seul scénario ?

**Réponse :**

Pour couvrir une large gamme de niveaux de charge cognitive car un seul scénario ne permettrait pas d'obtenir des exemples de faible et forte charge cognitive, et les modèles ne pourraient pas apprendre à discriminer. C'est aussi nécessaire pour que les modèles généralisent.

#### Q19. Pourquoi peut-on s’attendre à une charge cognitive plus élevée dans les scénarios 8 et 9 ?

### Réponse

Le scénario 8 demande en même temps de conduire et de suivre des instructions verbales pour réaliser un demi tour. Le scénario 9 impose une navigation dans un espace très contraint. Les deux mobilisent simultanément la coordination motrice fine, la mémoire de travail et l'attention visuelle (cad plusieurs ressources cognitives en même temps).

#### Q20. La complexité du scénario suffit-elle à définir le label de charge cognitive ?

### Réponse

Non et c'est pour ça que les auteurs utilisent les scores PAAS auto rapportés car 2 participants dans le même scénario peuvent avoir des charges cognitives très différentes selon leur niveau d'expertise, leur état du moment, ou leur familiarité avec les tâches. La complexité du scénario est un inducteur de charge mais pas un label en soi.


## 9. Échantillonnage et segmentation temporelle

Les signaux EEG sont continus. Pour entraîner un modèle, il faut transformer ce signal continu en exemples individuels.

Dans CL-Drive, les scores PAAS de charge cognitive sont collectés toutes les 10 secondes.

Il est donc naturel de découper les signaux physiologiques en fenêtres de 10 secondes :

$$
X_i = \text{segment de 10 s}
$$

$$
y_i = \text{score PAAS associé}
$$

On obtient ainsi le problème suivant :

Le principe d’association segment-label est le suivant :

| Temps | Segment EEG | Score PAAS |
|---|---|---|
| 0–10 s | segment 1 | score 1 |
| 10–20 s | segment 2 | score 2 |
| 20–30 s | segment 3 | score 3 |

## 10. Formulation du problème d'apprentissage automatique

À partir du dataset CL-Drive, on peut formuler plusieurs problèmes de classification.

### Classification binaire

| Scores PAAS | Classe |
|---|---|
| 1 à 4 | charge faible |
| 5 à 9 | charge élevée |

### Classification ternaire

| Scores PAAS | Classe |
|---|---|
| 1 à 3 | charge faible |
| 4 à 6 | charge moyenne |
| 7 à 9 | charge élevée |

Dans le projet, on commencera par la classification binaire.

#### Q27. Pourquoi commencer par une classification binaire plutôt qu’une classification ternaire ?

### Réponse

C'est plus simple car la frontière de décision est plus facile à apprendre, et les métriques d'évaluation sont plus directes. C'est aussi plus facile à interpréter pour valider que le pipeline fonctionne avant d'augmenter la difficulté

#### Q28. Dans une classification binaire, pourquoi le score PAAS 5 est-il généralement placé dans la classe « charge élevée » dans ce papier ?

### Réponse

Le score 5 correspond à "ni faible ni élevée" ce qui est concrètement déjà une charge non négligeable du point de vue de la sécurité routière donc il est plus prudent de l'inclure dans la classe à risque.

#### Q29. Quel est le lien entre ce papier et un système d'interface adaptative ?

**Réponse :**

L'idée est que si un véhicule peut détecter en temps réel que le conducteur est en surcharge cognitive, il peut adapter son comportement : réduire les notifications, suspendre la navigation vocale, alerter discrètement, ou proposer une assistance accrue. Donc le modèle entraîné sur CL Drive est la brique centrale de ce système d'adaptation.

## 9. Visualisation des signaux EEG

Avant tout prétraitement ou apprentissage automatique, il est indispensable de visualiser les signaux EEG.

La visualisation permet de :

- vérifier que les données sont correctement chargées ;
- observer les amplitudes ;
- repérer les valeurs aberrantes ;
- détecter les ruptures ou données manquantes ;
- comparer les canaux ;
- repérer des artefacts évidents ;
- comparer qualitativement des segments de faible et forte charge cognitive.

### Algorithme conceptuel pour visualiser un segment EEG

Pour visualiser un segment EEG de 10 secondes :

1. choisir un sujet ;
2. choisir un scénario ;
3. choisir un segment ;
4. récupérer les 4 canaux EEG ;
5. calculer le nombre d’échantillons : $N = f_s \times T$ ;
6. construire l’axe temporel : $t[n] = n/f_s$ ;
7. tracer les quatre canaux en fonction du temps ;
8. ajouter le score PAAS associé au titre du graphique ;
9. commenter la qualité du signal.

#### Q30. Que faut-il regarder en premier lorsqu’on affiche un signal EEG brut ?

### Réponse

L'amplitude globale (est elle réaliste? entre +100 µV ou -100 µV environ ?), les sauts brusques (artefacts de mouvement), les segments à zéro ou plats (données manquantes/électrode déconnectée), et les dérives lentes de la ligne de base.


#### Q31. Pourquoi afficher les quatre canaux EEG plutôt qu’un seul ?

### Réponse

Les canaux frontaux et temporaux capturent des régions différentes du cerveau. Afficher les 4 permet de repérer si un problème est local (un canal) ou global (tout le casque) parce qu'ils ne se comportent pas forcément de la même façon(un artefact peut toucher un seul canal).

#### Q32. Pourquoi comparer visuellement un segment de faible charge et un segment de forte charge cognitive ?

### Réponse

Pour vérifier que les signaux sont réellement différents selon le label. Si aucune différence n'est visible à l'œil entre un segment de charge 1 et un segment de charge 9, ça ne signifie pas qu'il n'y en a pas (l'EEG est complexe) mais ça encourage à vérifier la qualité des données et à réfléchir aux features à extraire. C'est aussi une façon d'acquérir une intuition sur le signal avant de passer au ML.

## 11. Pipeline général du projet

Le projet complet suivra la chaîne suivante :

$$
\text{EEG brut}
\rightarrow
\text{prétraitement}
\rightarrow
\text{segmentation}
\rightarrow
\text{features}
\rightarrow
\text{classification}
\rightarrow
\text{décision d'adaptation}
$$

Les TD suivants prépareront les étapes techniques : structure des données, prétraitement, extraction de caractéristiques, puis classification dans le TP projet.

#### Q33. Pourquoi l'utilisation d'un simulateur est-elle à la fois un avantage et une limite ?

**Réponse :**

Avantage car on contrôle parfaitement les conditions (scénarios reproductibles, sécurité totale, possibilité de varier la complexité de façon précise, pas de risque pour les participants). Limite puisque le comportement en simulateur n'est pas identique à la conduite réelle (les signaux physiologiques peuvent différer et le SAS introduit du bruit).
